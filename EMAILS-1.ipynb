```python
# Calculate correlation of each word with the target
print("=" * 50)
print("TOP 20 WORDS MOST CORRELATED WITH SPAM")
print("=" * 50)

correlations = []
for word in word_columns[:100]:  # Limit to first 100 words for performance
    corr = df[word].corr(df['Prediction'])
    correlations.append((word, corr))

# Sort by absolute correlation
correlations.sort(key=lambda x: abs(x[1]), reverse=True)

for i, (word, corr) in enumerate(correlations[:20], 1):
    print(f"{i:2d}. '{word:15s}' - Correlation: {corr:.4f}")

# Plot top correlations
top_corr_words = [w for w, _ in correlations[:15]]
top_corr_values = [c for _, c in correlations[:15]]

plt.figure(figsize=(12, 6))
colors = ['red' if c > 0 else 'blue' for c in top_corr_values]
plt.bar(range(len(top_corr_words)), top_corr_values, color=colors, alpha=0.7)
plt.xticks(range(len(top_corr_words)), top_corr_words, rotation=45, ha='right')
plt.xlabel('Words')
plt.ylabel('Correlation with Spam')
plt.title('Top 15 Words Correlated with Spam Classification')
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```
